```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = pd.read_csv("/content/CarPricesPrediction.csv")
print(data.head())
```

```python
print(data.info())
```

```python


#removing missing value
data_cleaned = data.dropna()
```

```python
# Removing columns with any missing values
data_cleaned = data.dropna(axis=1)
```

```python
# Histograms for 'Year', 'Mileage', and 'Price'
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
sns.histplot(data_cleaned['Year'], bins=30, kde=True)
plt.title('Distribution of Car Years')

plt.subplot(1, 3, 2)
sns.histplot(data_cleaned['Mileage'], bins=30, kde=True)
plt.title('Distribution of Car Mileage')

plt.subplot(1, 3, 3)
sns.histplot(data_cleaned['Price'], bins=30, kde=True)
plt.title('Distribution of Car Prices')

plt.tight_layout()
plt.show()
```

```python
# Box plots for 'Price' by 'Make' and 'Model' for categorical comparison
plt.figure(figsize=(12, 6))
sns.boxplot(x='Make', y='Price', data=data_cleaned)
plt.title('Car Prices by Make')
plt.show()

plt.figure(figsize=(14, 7))
sns.boxplot(x='Model', y='Price', data=data_cleaned)
plt.title('Car Prices by Model')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot for 'Price' vs. 'Mileage' and 'Year'
plt.figure(figsize=(18, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x='Mileage', y='Price', data=data_cleaned)
plt.title('Car Price vs. Mileage')

plt.subplot(1, 2, 2)
sns.scatterplot(x='Year', y='Price', data=data_cleaned)
plt.title('Car Price vs. Year')

plt.tight_layout()
plt.show()

desc_stats
```


The descriptive statistics and visualizations provide insights into the distribution of car prices and other relevant features in the dataset:

**Descriptive Statistics Summary:**

**Year:** The cars range from 2010 to 2022, with a mean year of approximately 2015.86, indicating a relatively recent selection of used cars.

**Mileage:** Mileage varies widely, from 10,079 to 149,794 miles, with a mean of around 78,796.93 miles. This wide range reflects the diverse usage history of the cars.

**Price:** Prices range from $12,613 to $31,414, with a mean price of approximately $22,195.21. The standard deviation is about $4,245.19, suggesting a moderate spread in car prices.

**Visualizations:**
